# 1 — Individual Trajectories Plot (Steps)

This plot shows the evolution of the number of steps across time for multiple participants.

Each thin line represents one participant.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("steps_data.csv")

# Create figure
plt.figure(figsize=(12,7))

# Plot each participant trajectory
for pid, g in df.groupby("PID"):
    
    plt.plot(
        g["day_index"],
        g["steps"],
        alpha=0.2
    )

plt.title("Daily Steps Trajectories")
plt.xlabel("Day Index")
plt.ylabel("Steps")
plt.grid(True)

plt.show()
```

# 2 — Mean Population Trend

This plot adds a thick line representing the **average behavior across participants**.


```python
df = pd.read_csv("steps_data.csv")

plt.figure(figsize=(12,7))

# Individual trajectories
for pid, g in df.groupby("PID"):
    
    plt.plot(
        g["day_index"],
        g["steps"],
        alpha=0.15
    )

# Mean trend
mean_trend = df.groupby("day_index")["steps"].mean()

plt.plot(
    mean_trend.index,
    mean_trend.values,
    linewidth=3,
    label="Mean Trend"
)

plt.title("Daily Steps Trajectories with Mean Trend")
plt.xlabel("Day Index")
plt.ylabel("Steps")
plt.legend()
plt.grid(True)

plt.show()
```

# 3 — Stacked Categorical Plot (Activity States)

This plot shows the **proportion of activity states across days**.

```python
df = pd.read_csv("activity_states.csv")

# Count activity states
daily = df.groupby(["day_index","activity_state"]).size().reset_index(name="count")

# Total per day
totals = daily.groupby("day_index")["count"].sum().reset_index(name="total")

daily = daily.merge(totals,on="day_index")

# Compute proportions
daily["proportion"] = daily["count"]/daily["total"]

# Convert to wide format
prop_wide = daily.pivot(
    index="day_index",
    columns="activity_state",
    values="proportion"
).fillna(0)

# Plot
prop_wide.plot(
    kind="bar",
    stacked=True,
    figsize=(16,8)
)

plt.ylabel("Proportion")
plt.xlabel("Day Index")
plt.title("Daily Activity State Composition")

plt.show()
```

# 4 — Sleep Density Heatmap

This plot visualizes where sleep onset and duration are most concentrated.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Load data
df = pd.read_csv("sleep_data.csv")

x = df["sleep_onset_hour"].values
y = df["sleep_duration_hr"].values

# Density estimation
values = np.vstack([x,y])
kde = gaussian_kde(values)

xgrid = np.linspace(20,28,300)
ygrid = np.linspace(3,12,300)

X,Y = np.meshgrid(xgrid,ygrid)

Z = kde(np.vstack([X.ravel(),Y.ravel()])).reshape(X.shape)

# Plot heatmap
plt.figure(figsize=(12,6))

plt.imshow(
    Z,
    origin="lower",
    aspect="auto",
    extent=[20,28,3,12],
    cmap="turbo"
)

plt.colorbar(label="Density")

plt.xlabel("Sleep Onset Time")
plt.ylabel("Sleep Duration (hours)")
plt.title("Sleep Onset vs Duration Density")

plt.show()
```
